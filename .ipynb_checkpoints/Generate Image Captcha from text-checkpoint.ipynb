# Generate Image Captcha in Python

```python
from captcha.image import ImageCaptcha
```

```python
#set the image size

image  = ImageCaptcha(width = 300, height = 100)

#Enter the text

cap_Text = input("Enter the Captcha Text : ")

#generate image
data = image.generate(cap_Text)

#save the image as file

image.write(cap_Text, 'cap1.png')
```

```python
#show the captcha of the text

from PIL import Image

Image.open('cap1.png')
```
